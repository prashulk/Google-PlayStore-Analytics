### Apps Cleaning

```python
import numpy as np
import pandas as pd
```

```python
g_data = pd.read_csv("apps.csv")
g_data.head()
```

```python
g_data.drop("Unnamed: 0", axis = 1, inplace = True)
g_data.head()
```

```python
g_data.info()
```

```python
g_data.isnull().sum()
```

```python
g_data.Price.value_counts()
```

```python
g_data.Price = g_data.Price.apply(lambda x: str(x).replace("$","") if "$" in str(x) else str(x))
g_data.Price = g_data.Price.apply(lambda x: float(x))
g_data.Price.value_counts()
```

```python
g_data.info()
```

```python
g_data.Installs.value_counts()
```

```python
g_data.Installs = g_data.Installs.apply(lambda x: str(x).replace("+","") if "+" in str(x) else str(x))
g_data.Installs = g_data.Installs.apply(lambda x: str(x).replace(",","") if "," in str(x) else str(x))
g_data.Installs = g_data.Installs.apply(lambda x: int(x))
g_data.Installs.value_counts()
```

```python
g_data.info()
```

```python
g_data.info()
```

```python
# Replace spaces with underscores in column names
g_data.columns = g_data.columns.str.replace(' ', '_')

# Display the modified DataFrame
g_data.info()
```

```python
import re
 
pattern = r'[^a-zA-Z0-9\s.,!?\':-]' 
 
mask1 = g_data['App'].str.contains(pattern, regex=True)
mask12 = g_data['Current_Ver'].str.contains(pattern, regex=True)

 
g_data2 = g_data[~(mask1 | mask12)]
g_data2.info()
```

```python
null_counts = g_data2.isnull().sum()
print("Null Counts:")
print(null_counts)

duplicates = g_data2.duplicated().sum()
print("\nNumber of Duplicates:", duplicates)

g_data2.info()
```

```python
g_data2 = g_data2.dropna()

g_data2.reset_index(drop=True, inplace=True)

g_data2.info()
```

```python
output_csv_path = 'cleaned_apps_f.csv'


# Write the DataFrame to a CSV file
g_data2.to_csv(output_csv_path, index=False)
```

### Reviews cleaning

```python
g_data = pd.read_csv("user_reviews.csv")
g_data.head()
```

```python
g_data.info()
```

```python
g_data.isnull().sum()
```

```python
null_counts = g_data.isnull().sum()
print("Null Counts:")
print(null_counts)

duplicates = g_data.duplicated().sum()
print("\nNumber of Duplicates:", duplicates)

g_data.info()
```

```python
g_data = g_data.dropna()

g_data.reset_index(drop=True, inplace=True)

g_data.info()
```

```python
import re
 
pattern = r'[^a-zA-Z0-9\s.,!?\':-]'  
mask1 = g_data['App'].str.contains(pattern, regex=True)
mask2 = g_data['Translated_Review'].str.contains(pattern, regex=True)
mask3 = g_data['Sentiment'].str.contains(pattern, regex=True)

g_data2 = g_data[~(mask1 | mask2 | mask3)]
g_data2.info()
```

```python
g_data2['Sentiment_Polarity'] = g_data2['Sentiment_Polarity'].round(2)
g_data2['Sentiment_Subjectivity'] = g_data2['Sentiment_Subjectivity'].round(2)

g_data2.head(20)
```

```python
g_data2.info()
```

```python
output_csv_path = 'cleaned_reviews_f.csv'


g_data2.to_csv(output_csv_path, index=False)
```
