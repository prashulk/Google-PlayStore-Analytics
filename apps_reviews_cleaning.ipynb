### Apps Cleaning

```python
import numpy as np
import pandas as pd
```

```python
g_data = pd.read_csv("apps.csv")
g_data.head()
```

```python
g_data.drop("Unnamed: 0", axis = 1, inplace = True)
g_data.head()
```

```python
g_data.info()
```

```python
g_data.isnull().sum()
```

```python
g_data.Price.value_counts()
```

```python
g_data.Price = g_data.Price.apply(lambda x: str(x).replace("$","") if "$" in str(x) else str(x))
g_data.Price = g_data.Price.apply(lambda x: float(x))
g_data.Price.value_counts()
```

```python
g_data.info()
```

```python
g_data.Installs.value_counts()
```

```python
g_data.Installs = g_data.Installs.apply(lambda x: str(x).replace("+","") if "+" in str(x) else str(x))
g_data.Installs = g_data.Installs.apply(lambda x: str(x).replace(",","") if "," in str(x) else str(x))
g_data.Installs = g_data.Installs.apply(lambda x: int(x))
g_data.Installs.value_counts()
```

```python
g_data.info()
```

```python
g_data.info()
```

```python
# Replace spaces with underscores in column names
g_data.columns = g_data.columns.str.replace(' ', '_')

# Display the modified DataFrame
g_data.info()
```

```python
import re
 
# Assuming you have a DataFrame called df
# Define a regular expression pattern to match the characters you want to remove
pattern = r'[^a-zA-Z0-9\s.,!?\':-]'  # This example pattern matches non-alphanumeric, non-space, non-punctuation characters
 
# Use the 'str.contains' method to create a boolean mask based on the pattern
mask1 = g_data['App'].str.contains(pattern, regex=True)
# mask2 = g_data['Category'].str.contains(pattern, regex=True)
# mask3 = g_data['Rating'].str.contains(pattern, regex=True)
# mask4 = g_data['Reviews'].str.contains(pattern, regex=True)
# mask5 = g_data['Size'].str.contains(pattern, regex=True)
# mask6 = g_data['Installs'].str.contains(pattern, regex=True)
# mask7 = g_data['Type'].str.contains(pattern, regex=True)
# mask8 = g_data['Price'].str.contains(pattern, regex=True)
# mask9 = g_data['Content_Rating'].str.contains(pattern, regex=True)
# mask10 = g_data['Genres'].str.contains(pattern, regex=True)
# mask11 = g_data['Last_Updated'].str.contains(pattern, regex=True)
mask12 = g_data['Current_Ver'].str.contains(pattern, regex=True)

 
# Invert the mask to keep rows that don't match the pattern
g_data2 = g_data[~(mask1 | mask12)]
g_data2.info()
```

```python
null_counts = g_data2.isnull().sum()
print("Null Counts:")
print(null_counts)

# Check for duplicates
duplicates = g_data2.duplicated().sum()
print("\nNumber of Duplicates:", duplicates)

# Display the modified DataFrame info
g_data2.info()
```

```python
g_data2 = g_data2.dropna()

# Reset the index if needed
g_data2.reset_index(drop=True, inplace=True)

# Display the modified DataFrame info
g_data2.info()
```

```python
output_csv_path = 'cleaned_apps_f.csv'


# Write the DataFrame to a CSV file
g_data2.to_csv(output_csv_path, index=False)
```

### Reviews cleaning

```python
g_data = pd.read_csv("user_reviews.csv")
g_data.head()
```

```python
g_data.drop("Unnamed: 0", axis = 1, inplace = True)
g_data.head()
```

```python
g_data.info()
```

```python
g_data.isnull().sum()
```

```python
null_counts = g_data.isnull().sum()
print("Null Counts:")
print(null_counts)

# Check for duplicates
duplicates = g_data.duplicated().sum()
print("\nNumber of Duplicates:", duplicates)

# Display the modified DataFrame info
g_data.info()
```

```python
g_data = g_data.dropna()

# Reset the index if needed
g_data.reset_index(drop=True, inplace=True)

# Display the modified DataFrame info
g_data.info()
```

```python
import re
 
# Assuming you have a DataFrame called df
# Define a regular expression pattern to match the characters you want to remove
pattern = r'[^a-zA-Z0-9\s.,!?\':-]'  # This example pattern matches non-alphanumeric, non-space, non-punctuation characters
 
# Use the 'str.contains' method to create a boolean mask based on the pattern
mask1 = g_data['App'].str.contains(pattern, regex=True)
mask2 = g_data['Translated_Review'].str.contains(pattern, regex=True)
mask3 = g_data['Sentiment'].str.contains(pattern, regex=True)
# mask4 = g_data['Sentiment_Polarity'].str.contains(pattern, regex=True)
# mask5 = g_data['Sentiment_Subjectivity'].str.contains(pattern, regex=True)
# # mask6 = g_data['Installs'].str.contains(pattern, regex=True)
# # mask7 = g_data['Type'].str.contains(pattern, regex=True)
# # mask8 = g_data['Price'].str.contains(pattern, regex=True)
# # mask9 = g_data['Content_Rating'].str.contains(pattern, regex=True)
# # mask10 = g_data['Genres'].str.contains(pattern, regex=True)
# # mask11 = g_data['Last_Updated'].str.contains(pattern, regex=True)
# mask12 = g_data['Current_Ver'].str.contains(pattern, regex=True)

 
# Invert the mask to keep rows that don't match the pattern
g_data2 = g_data[~(mask1 | mask2 | mask3)]
g_data2.info()
```

```python
# Assuming 'df' is your DataFrame
g_data2['Sentiment_Polarity'] = g_data2['Sentiment_Polarity'].round(2)
g_data2['Sentiment_Subjectivity'] = g_data2['Sentiment_Subjectivity'].round(2)

g_data2.head(20)
```

```python
g_data2.info()
```

```python
output_csv_path = 'cleaned_reviews_f.csv'


# Write the DataFrame to a CSV file
g_data2.to_csv(output_csv_path, index=False)
```
